```python
def jacard(set1, set2):
    set1_list = set()
    set2_list = set()
    
    for im_name, pred in set1.items():
        add = (im_name, pred)
        set1_list.add(add)
#         set1_list.add(im_name+str(pred))
    for im_name, pred in set2.items():
        add = (im_name, pred)
        set2_list.add(add)
#         set2_list.add(im_name+str(pred))
    
    inter = float(len(set1_list.intersection(set2_list)))
    union = float(len(set1_list.union(set2_list)))
    
    return inter/union
```
